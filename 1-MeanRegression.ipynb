```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
N = 50
data = 2*np.random.rand(N) + 0.5 + 0.1*np.random.randn(N)
plt.scatter(data, 0*data)
```

```python
def loss(data, mu):
    return np.sum((data-mu)**2)
test_mus = np.linspace(0, 3)
err = np.array([loss(data, mu) for mu in test_mus])
plt.scatter(data, 0*data)
plt.plot(test_mus, err)
```
